## Plotting exercise 

### 1. create a scatter plot of the total tests per country versus the population
### 2. create a cratter plot of total test per country versus the precentage of cases
### 3. create a scatter plot of % of tests per country veersus % of total cases for each country


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# using last sheet from the excel file , as the most up to date data 

df = pd.read_excel('Coronavirus_Data.xlsx',sheet_name = '14-6',skipfooter =1)
df.sample(5)
```

# 1

```python
total_tests_1000s = df['Total Tests'] /1000
pop = df['Population']/1_000_000
```

```python
plt.figure(figsize = (10,6))
plt.style.use('ggplot')
plt.title ('Total tests per country versus the population',fontsize = 20, color = 'Red')
plt.xlabel("Total tests in 1000's " ,fontsize = 20, color = 'Blue')
plt.ylabel ('Population in Milliions',fontsize = 20, color = 'Blue')

plt.scatter (total_tests_1000s,pop)
plt.show()
```

# 2

```python
#  add new column to show the % of cases per coutry population 

df['Percentage_0f_cases_per_pop'] = df["Total Cases"] / df["Population"] * 100
total_tests_1000s = df['Total Tests'] /1000
pcpp = df['Percentage_0f_cases_per_pop']

plt.figure(figsize = (10,6))
plt.style.use('ggplot')

plt.title ('Total tests per country versus the percentage of cases',fontsize = 20, color = 'Red')
plt.xlabel("Total tests in 1000's " ,fontsize = 20, color = 'Magenta')
plt.ylabel ('Peercentage cases',fontsize = 20, color = 'Magenta')

plt.scatter (total_tests_1000s,pcpp)
plt.show()
```
