```python
import matplotlib.pyplot as plt
import pandas as pd
import random
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')

%matplotlib inline
```

```python
x =[x for x in range(1,11)]
y1 = [3, 6, 1, 11, 9, 5, 13, 15, 17, 19]
y2 =[1,6,7,11,13,15,16,23,25,30]
print(x)
print(y1)
print(y2)
```

```python
# my facecolor must be some color cos of the dark mode
plt.figure(facecolor= 'yellow')
plt.title ('Sample Plot1')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.plot(x,y1,label ='y1')
plt.plot(x,y2,label='y2')

plt.legend()
plt.show()
```

```python
# create the same plot using a slightly diferent notation AXES
fig,axes = plt.subplots(facecolor = 'g')

#now add some data to the plot
axes.plot(x,y1,label = 'y1')
axes.plot(x,y2,label = 'y2')

#set the facecolor :( not working 
axes.set_facecolor('pink')

axes.set_title('Sample Plot2')
axes.set_xlabel('X Values')
axes.set_ylabel('Y Values')

axes.legend()
#set the title of the 'axes' (plot)
axes.set_title('Sample plot')
plt.savefig('Test.png')
plt.show()
```

```python
# creating plot with 1 row 2 columns which will give us two plots/axes side by side
# subplots(rows,columns)
fig,axes_list = plt.subplots(1,2,facecolor = 'y')
axes_list[0].plot(x,y1,color='green')
axes_list[1].plot(x,y2,color='magenta')
plt.show()
```

```python
# Creating  plot with 1 row and two columns which will give us two plots/axes side by side

# subplots(rows, columns)
# We can also set the figure size when creating the figure 'figisize=(width, height) in inches'
fig, axes_list = plt.subplots(1,2, figsize=(8,4),facecolor = 'y')

# plot soe data on the two axes that we have and set the label names
axes_list[0].plot(x,y1, color="magenta", label="y1")
axes_list[1].plot(x,y2, color="cyan", label="Y2")

# Add padding to figure to increase space between the plots
fig.tight_layout(pad=3)

# Set the labels and titles for the plots
axes_list[0].set_xlabel("X values")
axes_list[1].set_xlabel("X values")

axes_list[0].set_ylabel("Y1 values")
axes_list[1].set_ylabel("Y2 values")

axes_list[0].set_title("Sub Plot A")
axes_list[1].set_title("Sub Plot B")

# Show the legends
axes_list[0].legend(["Tada!"],fontsize="14", loc="lower right")
axes_list[1].legend()

fig.savefig("Subplpots Example") # Save the fig

plt.show()
```
