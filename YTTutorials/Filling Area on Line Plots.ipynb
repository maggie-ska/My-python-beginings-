```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data1.csv',delimiter=';')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

#plt.fill_between 
# for the moment its filling from zero to python line but we can use median salary 
overall_median = 57287

# to fill the area under the python line this will block the view a little bitso we specify ALPHA 
# we are passing the third argument the median salary 
plt.fill_between(ages,py_salaries,overall_median , where = (py_salaries > overall_median), alpha = 0.25)

#where = (py_salaries > overall_median) - conditional if we create a second plt fill between we get two colors 
plt.fill_between(ages,py_salaries,overall_median , where = (py_salaries <= overall_median), alpha = 0.25) # color = 'red'


plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.legend()
plt.show()
```
