## **<font color = 'orange'>tfidf for the 3 documents**

```python
# Function to convert/split text to words
def text_to_words(text):
    import re
    # [12], [1], [123] <- Wiki citation char sequences
    # \[\d+\]
    # 's
    text = re.sub("\(|\)|\[\d+\]|('s)", "", text)
    # Match a single space " " OR ", " (comma and a space)
    split_pattern = " |, |-|\n|\."
    text = re.split(split_pattern, text)    
    # Only keep terms that are alphabetic characters
    text = [word for word in text if word.isalpha()]
    print(f"The text has been converted to {len(text)} words.")
    
    return text
```

```python
def read_all_text(filename):
    with open(filename, encoding="utf-8") as f:
        print(f"\n\nReading contents of file: {filename}")
        return f.read().lower()

spacex_text = read_all_text("spacex.txt")
spacex_words = text_to_words(spacex_text)

# Get all the words for the blue origin and virgin galactic files
blue_origin_text = read_all_text("blue-origin.txt")
blue_origin_words = text_to_words(blue_origin_text)

virgin_galactic_text = read_all_text("virgin-galactic.txt")
virgin_galactic_words = text_to_words(virgin_galactic_text)
```

**<font color = 'orange'> do the tf for the words**

```python
def frequency (words):
    unique_words = list(set(words))
    
    word_freq = {}

    for word in words:
        count = word_freq.get(word, 0)
        word_freq[word] = count + 1
    return word_freq   
        
```

```python
freq_spacex = frequency(spacex_words) 
freq_blue_o = frequency(blue_origin_words) 
freq_virgin_g = frequency(virgin_galactic_words) 
#freq_spacex
```

**<font color = 'orange'> making dataframes**

**<font color = 'magenta'> spacex document**

```python
import pandas as pd

df_spacex = pd.DataFrame(freq_spacex.items(), columns = ["Word","Frequency"])

# Add a column
df_spacex["Document Name"] = "Spacex.txt"

# Change the order of the columns
df_spacex = df_spacex[["Document Name","Word","Frequency"]]

spacex_word_count = len(spacex_words)

# Add the TF (Term Frequency) column
# TF = (number of occurances of term) / (total number of terms in the doc)
df_spacex["TF"] = df_spacex["Frequency"]/spacex_word_count

# adding Document Frequency column 
doc_freqs = []
# Looping through the words from the df (which ensures that my counts will also be in the correct order)
for word in df_spacex["Word"]:
    doc_freq_count = 0 
    if word in spacex_words:
        doc_freq_count += 1
    if word in virgin_galactic_words:
        doc_freq_count += 1
    if word in blue_origin_words:
        doc_freq_count += 1
    # Add the doc frequency count of the that word to the list
    doc_freqs.append(doc_freq_count)

# Add the DF colum and values to 
df_spacex["DF"] = doc_freqs

# adding the last column 
import math

# Lsit to hold calcualted TF-IDF values for each word in the dataframe
tf_idf_spacex = []

# Our collection has 3 docs 
num_docs_in_corpus = 3 

for row in df_spacex.values:
    # Get the current rows TF value
    tf = row[3]
    # Get the current rows DF value
    doc_freq = row[4]   
    # Calculate the TF-IDF value
    tf_idf = tf * math.log(num_docs_in_corpus / doc_freq)
    # Add that calculated value ourt list
    tf_idf_spacex.append(tf_idf)

# Add out Calculated TF-IDF values to our datafraame as a new column
df_spacex["TF-IDF"] = tf_idf_spacex

df_spacex.sort_values("TF-IDF", ascending=False,inplace=True, ignore_index=True)
df_spacex.head(10)
    
```
