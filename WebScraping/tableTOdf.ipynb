### scrape table example 1 

```python
# uniform resource locator
url = 'https://keithgalli.github.io/web-scraping/webpage.html'
```

```python
import requests
from bs4 import BeautifulSoup as bs 
import pandas as pd
import numpy as np
```

```python
response = requests.get(url)
response
```

```python
web_page = response.content
soup = bs(web_page)
#soup
```

### need to check on the website how to grab the table 

```python
# because its only one table on the website we can just look for table 
table_soup = soup.find("table")
table_soup

# just to be sure we are grabbing the right table we can add the class
table_soup = soup.find("table",attrs= {"class":"hockey-stats"})
table_soup

# pd.read_html returns a list of dataframes 
dfs = pd.read_html(str(table_soup))

# the first (and only df will be the table we got out of the soup object )
dfs[0]
table1 = dfs[0]
table1
```

```python
# we dont even need a beautifulsoup -> we can just use Pandas -> this wont always work 

dfs = pd.read_html(url)
table1 = dfs[0]

# sometimes we will need this code or soup 
dfs = pd.read_html(requests.get(url).text)
dfs[0]
```
