### 1. using the spacex.txt file perform a textual word frequency analysis


* show the frequency of occurrence of all the words in the file 
* make a bar plot to show the top 15 most frequently occuring words
* make a bar plot to show the bottom 15 least frequently occuring words


```python
# first we need to read in the file, what will i use hmmmm
import pandas as pd 



df = pd.read_table('spacex.txt')   
df.head()
```

```python
import numpy as np 

with open ('spacex.txt') as f :
    words = f.read().split()
    print(words)
```

```python
def word_count(text):
    """returns a dictionary with the number of times each word occurs """
    word_frequency_dict = dict()
    word_list = text

    for word in word_list:
            
	    if word in word_frequency_dict:
	        word_frequency_dict[word] += 1
	    else:
	        word_frequency_dict[word] = 1

    return word_frequency_dict
```

```python
list_of_words = word_count(words)
len(list_of_words)
```

```python
#df = pd.Series(word_frequency_dict).to_frame('Frequency')

df = pd.DataFrame.from_dict(list_of_words ,orient='index' )



df.columns = ['count']
df.sort_values('count',ascending = False,inplace = True)
df.index.name ='word'
df.head(15)
```

```python
# words that occured one time only 
df[df['count']==1]
```

```python
top15 = df[:15]
axes = top15.plot(kind = 'bar', figsize =(8,6))

# annotate the bars in the axes 
for patch in axes.patches:
    axes.annotate(str(patch.get_height()),
                 ((patch.get_x()+patch.get_width()/2),patch.get_height()),
                 ha= 'center',va='bottom')
```
