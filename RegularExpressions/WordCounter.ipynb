### 1. using the spacex.txt file perform a textual word frequency analysis


* show the frequency of occurrence of all the words in the file 
* make a bar plot to show the top 15 most frequently occuring words
* make a bar plot to show the bottom 15 least frequently occuring words


```python
# first we need to read in the file, what will i use hmmmm
import pandas as pd 
import numpy as np 

with open ('spacex.txt') as f :
    words = f.read().split()   


# checking for unique words and counting them 
np.unique(words,return_counts = True) 

# unpacking the two values
uWords,count = np.unique(words,return_counts = True)
uWords
```

```python
for word,freq in zip(uWords,count):
    print(f'{word} : {freq}')
```

```python
data=pd.DataFrame(uWords,count)
data.sort_values(0).tail(15)
```
