```python
# twitter api test 2
import keys, tweepy
auth = tweepy.OAuthHandler(keys.consumer_key,keys.consumer_key_secret)
auth.set_access_token(keys.access_token,keys.access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
```

```python
query= input("What do you what to search twitter for?")
print("Gettng tweets for sentiment analysis...\n")
tweets = api.search(q=query, count=20, lang="en", tweet_mode="extended")
print(f"{len(tweets)} were found.")
from textblob import Blobber
from textblob.sentiments import NaiveBayesAnalyzer
# Bloober:
# A factory for TextBlobs that all share the same tagger,
# tokenizer, parser, classifier, and np_extractor.

factory_blobber = Blobber(analyzer=NaiveBayesAnalyzer())

tweet_count = 1
full_text_count = 0

for tweet in tweets:
    try:
        tweet_text = tweet.full_text
        full_text_count += 1          
    except:
        tweet_text = tweet.text
               
    print(f"\nTWEET NUMBER: {tweet_count}")
    print(f"Twitter user ID: {tweet.id}")
    print(f"{tweet_text}")
    print(f"{'Sentiment Analysis Classification'.upper()}: {factory_blobber(tweet_text).sentiment.classification.upper()}")
    tweet_count += 1
    
print("\nFINISHED!")
print(f"{full_text_count}")
```
