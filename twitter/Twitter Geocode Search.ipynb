```python
import tweepy,utils
from geopy.geocoders import Nominatim
from functools import partial

api = utils.create_tweepy_api()
```

```python
query = 'pis'

# the search function can take another parameter -GEOCODE lat,lon,radus with km(kilometers) or mi(miles) 
tweets = api.search(q=query,lang='pl',tweet_mode='extended',geocode =' 51.3974465,16.2051275,20km')
len(tweets)
```

### <font color = 'orange'>use the geolocation lat/long and radius to make a tweepy geocode string 

```python
loc = input('Location of search: ')
rad = input("What's the radius for the seach in km ??")
query2 = input ('Query to search for in the twitter?')
#en es,fr,pl look up language codes 

geolocator = Nominatim(user_agent ='Test app')
loc_object = geolocator.geocode(loc,language ='en')

print(f'This is the result for search: {query2} , in: {loc_object.address}')
lat = loc_object.latitude
long =loc_object.longitude
rad = rad+'km'
location_str = str(lat)+','+str(long)+','+rad

# or 
# location_str = f"{loc_object.latitude},{loc_object.longitude}{rad}km"

tweets2 = api.search(q=query2,lang='en',tweet_mode='extended',geocode =location_str)

df2 = utils.dataframe_from_tweets(tweets2)
df2
```
