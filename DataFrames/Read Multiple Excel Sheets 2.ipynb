#### Read Mutliple Excel sheets in to a single DataFrame Dictionary

```python
import pandas as pd
```

```python
xls = pd.ExcelFile("Coronavirus data.xlsx")
```

```python
# View the sheets in the Excel file
xls.sheet_names
```

```python
# Import the first the sheet only
first_sheet = pd.read_excel(xls, header=[0,1])
```

```python
first_sheet
```

```python
# Read the sheet called '20-3'
df_20_3 = pd.read_excel(xls,sheet_name="20-3")
df_20_3
```

```python
# Read multiple sheets at once as dictionary of Dataframes
multple_sheets_1 = pd.read_excel(xls,sheet_name=["20-3","21-3","22-3","23-3","24-3"]) # Can use list of int's or strings
```

```python
type(multple_sheets_1)
```

```python
multple_sheets_1 # {"key1" : df1 , "key2" : df2, .......}
```

```python
# Show one of the df's in the dictionary
multple_sheets_1["20-3"]
```

```python
# Show the keys of the dictionary
multple_sheets_1.keys()
```

```python
# Show the Total cases from the df for the key 23-3 (Using the dictionary we just made!)
# multple_sheets_1["23-3"] is a DataFrame

total_cases_23_3 = multple_sheets_1["23-3"]["Total Cases"] # df["Column Name"] which returns a Pandas series!
total_cases_23_3

# Then get the sum ,min and max values of the total cases from above
print(type(total_cases_23_3))

print("Sum of total cases: ", total_cases_23_3.sum() )
print("Min of total cases: ", total_cases_23_3.min() )
print("Max of total cases: ", total_cases_23_3.max() )
```

```python
# We can use descibe() on any series
# Get the column of New Cases in the 24-3 df from the dictionary
multple_sheets_1["24-3"]["New Cases"].describe()
```

```python
# Read all of the sheets in the excel file at once
# To read all sheets supply the parameter of sheet_name=None
all_sheets = pd.read_excel(xls,sheet_name=None)
```

```python
all_sheets # <- Is a dictionary of dataframes of all the sheets on the Excel File
```

```python
all_sheets.keys()
```

```python
xls.sheet_names # The keys in the dictionary are the same as the sheet names.
```

```python
# Get the index of the value "1-6" from the list xls.sheet_names
index_1_6 = xls.sheet_names.index("1-6")
index_1_6
```

```python
# Get all the sheets from the 1-6 up to the last sheet
# xls.sheet_names is a list of the sheet name. Use slicing

dict_june_dfs = pd.read_excel(xls, sheet_name=xls.sheet_names[index_1_6:])
dict_june_dfs
```
