```python
# Using the coronaviurs excel file
# Read the sheets, 14-3, 15-3,16-3 in to df's
# Set the Country column as the index for the df's
# Then get the data for the USA from each df

import pandas as pd

xls = pd.ExcelFile("Coronavirus data.xlsx")

# usecols to not include any extra row in the excel sheet
# We skipped the first row for good measure
df_14_3 = pd.read_excel(xls, sheet_name="14-3", index_col=0 , usecols="A:J", skiprows=1)
```

```python
df_15_3 = pd.read_excel(xls, sheet_name="15-3", index_col=0 , usecols="A:J", skiprows=1)
df_15_3
```

```python
df_16_3 = pd.read_excel(xls, sheet_name="16-3", index_col=0 , usecols="A:J", skiprows=1)
df_16_3
```

```python
usa_data1 = df_14_3.loc["USA"]
usa_data1 # Series with the data from the row with index "USA"
```

```python
usa_data2 = df_15_3.loc["USA"]
usa_data2 # Series with the data from the row with index "USA"
```

```python
usa_data3 = df_16_3.loc["USA"]
usa_data3 # Series with the data from the row with index "USA"
```

```python
# BONUS: Combine the USA data from each series into a single dataframe
df_usa = pd.DataFrame([usa_data1,usa_data2,usa_data3])
df_usa
```

```python
# Reset the indexes to defualt numeric indexes
# Must use the inplace=True parameter to change the actual dataframe
# (Other a new dataframe is returned)
df_usa.reset_index(inplace=True)
df_usa
```

```python
# Lets rename the columns to more suitbale names
new_names ={
    "index":"Country",
    "Date": "Date",
    "Cases" : "Total Cases",
    "Cases.1": "New Cases",
    "Deaths" : "Total Deaths",
    "Deaths.1" : "New Deaths",
    "Recovered" : "Total Recovered",
    "Cases.2" : "Active Cases",
    "Critical" : "Serious / Critical",
    "1M pop" : "Total Cases / 1M pop"
}
df_usa.rename(columns=new_names, inplace=True)
df_usa
```

```python
# Change the date in row one of the df_usa to 2020-03-15
# df.at[] or df.iat[] = We can use these to get or set a value in cells in a dataframe

# Set the values of the row 1 Date column cell to a Timestamp (datetime) object that represents 
# the date 15-03-2020
# pandas.Timestamp() is equivalent to the standard Python datetime type from the datetime library. 
df_usa.at[1,"Date"] = pd.Timestamp(year=2020, month=3, day=15)
df_usa
```

```python
# Make a plot of the Date (x values) against the Total Cases, Total Deaths.
# Use two different colored lines
# HINT: Get columns as a series.

dates = df_usa["Date"] # x values
total_cases = df_usa["Total Cases"] # y values
total_deaths = df_usa["Total Deaths"] # y values

# Lets add the new cases
plt.plot(dates,total_cases, color="y", marker="o") # First line plotted
plt.plot(dates,total_deaths, color="r", marker="o") # Second line plotted
plt.plot(dates, df_usa["New Cases"],color="b", marker="o") # Third libe plotted

# We can add a legend to this plot in to indicate what each line represents
# loc can be things like
# 'upper left', 'upper right', 'lower left', 'lower right' 
plt.legend(['Total Cases','Total Deaths', 'New Cases'],loc="upper left")

# Set the xticks after plotting to the dates that we are using from
# our at Date Column
plt.xticks(ticks=dates, rotation=45)

plt.ylabel("Cases")
plt.xlabel("Date")

plt.show()
```
