```python
# Using the coronaviurs excel file
# Read the sheets, 14-3, 15-3,16-3 in to df's
# Set the Country column as the index for the df's
# Then get the data for the USA from each df

import pandas as pd

xls = pd.ExcelFile("Coronavirus_Data.xlsx")

# usecols to not include any extra row in the excel sheet
# We skipped the first row for good measure
df_14_3 = pd.read_excel(xls, sheet_name="14-3", index_col=0 , usecols="A:J", skiprows=1)
```

```python
df_15_3 = pd.read_excel(xls, sheet_name="15-3", index_col=0 , usecols="A:J", skiprows=1)
df_15_3
```

```python
df_16_3 = pd.read_excel(xls, sheet_name="16-3", index_col=0 , usecols="A:J", skiprows=1)
df_16_3
```

```python
usa_data1 = df_14_3.loc["USA"]
usa_data1 # Series with the data from the row with index "USA"
```

```python
usa_data2 = df_15_3.loc["USA"]
usa_data2 # Series with the data from the row with index "USA"
```

```python
usa_data3 = df_16_3.loc["USA"]
usa_data3 # Series with the data from the row with index "USA"
```

```python
# BONUS: Combine the USA data from each series into a single dataframe
df_usa = pd.DataFrame([usa_data1,usa_data2,usa_data3])
df_usa
```

```python
# Reset the indexes to defualt numeric indexes
# Must use the inplace=True parameter to change the actual dataframe
# (Other a new dataframe is returned)
df_usa.reset_index(inplace=True)
```

```python
# Lets rename the columns to more suitbale names
new_names ={
    "index":"Country",
    "Date": "Date",
    "Cases" : "Total Cases",
    "Cases.1": "New Cases",
    "Deaths" : "Total Deaths",
    "Deaths.1" : "New Deaths",
    "Recovered" : "Total Recovered",
    "Cases.2" : "Active Cases",
    "Critical" : "Serious / Critical",
    "1M pop" : "Total Cases / 1M pop"
}
df_usa = df_usa.rename(columns=new_names)
df_usa
```

```python

%matplotlib inline
d = df_usa['Date']
#date = ['14.3.20','15.3.20','16.3.20']
NewCases = df_usa['New Cases']
NewDeaths = df_usa['New Deaths']
plt.style.use('seaborn-ticks')


#plt.figure(figsize=(14,8))
plt.plot(d,NewCases,color = 'red',linestyle =':',linewidth = 2,marker = 'd',label = 'New Cases')
plt.plot(d,NewDeaths,color = 'blue',linewidth = 2,label = 'New Deaths')

#plt.plot(date,totalDeaths,color = '#fc4f30',label = 'Total Deaths')
plt.title ('USA covid 19 stats for 3 days in March',size = 14,fontweight ='500',color = 'magenta')
plt.xlabel('Date')
plt.ylabel('Cases')

plt.legend(loc= 'upper left')
plt.xticks (ticks = d)

#plt.xscale('log')
#plt.yscale('log')
plt.grid(True)
plt.show()
```
